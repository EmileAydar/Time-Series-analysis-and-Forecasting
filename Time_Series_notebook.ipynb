### DPAE

####  Deux séries temporelles en fonction du contrat (CDD ou CDI)
- DPAE (total) = DPAE (brut) + DPAE (cvs)

```python

# La troisième colonne contient des symboles inconnus, nous utilisons donc ici des nombres au lieu de noms de colonnes
dpae = dpae.sort_values(by=dpae.columns[2])

# Extraire les données CDD et CDI
dpae_cdd = dpae[dpae['Nature de contrat'] == "CDD de plus d'un mois"]
dpae_cdi = dpae[dpae['Nature de contrat'] == "CDI"]

# Mise en place des index
dpae_cdd = dpae_cdd.set_index(pd.to_datetime(dpae_cdd.iloc[:, 2]))
dpae_cdi = dpae_cdi.set_index(pd.to_datetime(dpae_cdi.iloc[:, 2]))

# définir la date comme index
dpae['DPAE (total)'] = dpae['DPAE (brut)'] + dpae['DPAE (cvs)']

dpae_cdd['DPAE (total)'] = dpae_cdd['DPAE (brut)'] + dpae_cdd['DPAE (cvs)']
dpae_cdi['DPAE (total)'] = dpae_cdi['DPAE (brut)'] + dpae_cdi['DPAE (cvs)']
series_cdd = dpae_cdd['DPAE (total)']
series_cdi = dpae_cdi['DPAE (total)']

# dessiner un graphique de série chronologique
plt.figure(figsize=(12, 6))
plt.plot(series_cdd, label='CDD de plus d\'un mois')
plt.plot(series_cdi, label='CDI')
plt.xlabel('Date')
plt.ylabel('DPAE')
plt.title('CDD & CDI')
plt.legend()
plt.show()

```

#### Les statistiques classiques : moyennes, médianes, écart type, variance
#### Des boîtes à moustaches

```python


# rééchantillonner les données pour la fréquence annuelle et trimestrielle et calculer la somme
# DPAE CDD YEAR
dpae_cdd_year = dpae_cdd['DPAE (total)'].resample('Y').sum()
dpae_cdd_year.index = dpae_cdd_year.index.year
dpae_cdd_year.index.name = 'Year'
# DPAE CDI YEAR
dpae_cdi_year = dpae_cdi['DPAE (total)'].resample('Y').sum()
dpae_cdi_year.index = dpae_cdi_year.index.year
dpae_cdi_year.index.name = 'Year'

dpae_cdd_quarter = dpae_cdd['DPAE (total)'].resample('Q').sum()
dpae_cdd_quarter.index = dpae_cdd_quarter.index.to_period('Q').astype(str)
dpae_cdd_quarter.index.name = 'Quarter'

dpae_cdi_quarter = dpae_cdi['DPAE (total)'].resample('Q').sum()
dpae_cdi_quarter.index = dpae_cdi_quarter.index.to_period('Q').astype(str)
dpae_cdi_quarter.index.name = 'Quarter'


cdd_year_stats = pd.DataFrame({
    'mean': dpae_cdd_year.mean(),
    'median': dpae_cdd_year.median(),
    'std': dpae_cdd_year.std(),
    'var': dpae_cdd_year.var()
}, index=['CDD_year'])

cdi_year_stats = pd.DataFrame({
    'mean': dpae_cdi_year.mean(),
    'median': dpae_cdi_year.median(),
    'std': dpae_cdi_year.std(),
    'var': dpae_cdi_year.var()
}, index=['CDI_year'])

# Calculer les statistiques trimestrielles CDI
cdi_quarter_stats = pd.DataFrame({
    'mean': dpae_cdi_quarter.mean(),
    'median': dpae_cdi_quarter.median(),
    'std': dpae_cdi_quarter.std(),
    'var': dpae_cdi_quarter.var()
}, index=['CDI_quarter'])

# Calculer les statistiques trimestrielles CDD
cdd_quarter_stats = pd.DataFrame({
    'mean': dpae_cdd_quarter.mean(),
    'median': dpae_cdd_quarter.median(),
    'std': dpae_cdd_quarter.std(),
    'var': dpae_cdd_quarter.var()
}, index=['CDD_quarter'])

data_summary = pd.concat([cdd_year_stats, cdd_quarter_stats, cdi_year_stats, cdi_quarter_stats])
data_summary

# Dessiner une boîte à moustaches
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].boxplot([dpae_cdd_year, dpae_cdi_year], labels=['CDD YEAR', 'CDI YEAR'])
ax[0].set_title('DATA BY YEAR')
ax[0].set_ylabel('DPAE')

ax[1].boxplot([dpae_cdd_quarter, dpae_cdi_quarter], labels=['CDD QUARTER', 'CDI QUARTER'])
ax[1].set_title('DATA BY QUARTER')
ax[1].set_ylabel('DPAE')

plt.show()

```

### Question 3 : prédiction par lissage exponentiel
Séparer vos données en ensemble d’apprentissage et de test pour effectuer des prédictions par lissage 
exponentiel sur les données DPAE (CDD et CDI) et EESSA (secteurs sélectionnés). 
#### Prédictions sur les données DPAE (CDD et CDI) par None seasonal Holt-Winters exponential smoothing

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

data_cdd = dpae_cdd['DPAE (total)']
data_cdi = dpae_cdi['DPAE (total)']

# Diviser l'ensemble de données en ensembles d'apprentissage et de test (fractionnement à 80 %)
train_size = int(len(data_cdd) * 0.8)
train_cdd, test_cdd = data_cdd[:train_size], data_cdd[train_size:]
train_cdi, test_cdi = data_cdi[:train_size], data_cdi[train_size:]

# Application de la méthode de lissage exponentiel de Holt-Winters aux prévisions
fit_cdd = Holt(train_cdd, initialization_method="estimated").fit(smoothing_level=0.2, smoothing_trend=0.2)
fit_cdi = Holt(train_cdi, initialization_method="estimated").fit(smoothing_level=0.2, smoothing_trend=0.2)

# Faire des prédictions sur l'ensemble de test
n_forecast = 10
forecast_cdd = fit_cdd.forecast(len(test_cdd) + n_forecast)
forecast_cdi = fit_cdi.forecast(len(test_cdi) + n_forecast)

# Calculer la courbe complète (y compris les ajustements de l'ensemble d'entraînement et les prédictions de l'ensemble de test）
full_curve_cdd = np.concatenate((fit_cdd.fittedvalues, forecast_cdd))
full_curve_cdi = np.concatenate((fit_cdi.fittedvalues, forecast_cdi))


plt.figure(figsize=(12, 6))
plt.plot(data_cdd.values, marker="o", color="black", label='CDD')
plt.plot(full_curve_cdd, marker="o", color="orange", label='CDD Forecast')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(data_cdi.values, marker="o", color="black", label='CDI')
plt.plot(full_curve_cdi, marker="o", color="blue", label='CDI Forecast')
plt.legend()
plt.show()
```

Utiliser d’autres modèles (par exemple ARMA) pour effectuer la prédiction. Comparer les résultats. 

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Entraîner les modèles ARIMA
model_cdd_arma = ARIMA(train_cdd, order=(2, 0, 2))
model_cdi_arma = ARIMA(train_cdi, order=(2, 0, 2))
model_sec1_arma = ARIMA(train_sec1, order=(2, 0, 2))
model_sec2_arma = ARIMA(train_sec2, order=(2, 0, 2))
model_sec3_arma = ARIMA(train_sec3, order=(2, 0, 2))

# On fit les modèles aux données d'entraînement
model_cdd_arma_fit = model_cdd_arma.fit()
model_cdi_arma_fit = model_cdi_arma.fit()
model_sec1_arma_fit = model_sec1_arma.fit()
model_sec2_arma_fit = model_sec2_arma.fit()
model_sec3_arma_fit = model_sec3_arma.fit()

# On effectue des prédictions sur les données test
forecast_cdd_arma = model_cdd_arma_fit.forecast(steps=len(test_cdd))
forecast_cdi_arma = model_cdi_arma_fit.forecast(steps=len(test_cdi))
forecast_sec1_arma = model_sec1_arma_fit.forecast(steps=len(test_sec1))
forecast_sec2_arma = model_sec2_arma_fit.forecast(steps=len(test_sec2))
forecast_sec3_arma = model_sec3_arma_fit.forecast(steps=len(test_sec3))

# On calcule le MSE et le RMSE pour les contrats CDI et CDD
mse_cdd = mean_squared_error(test_cdd, forecast_cdd_arma)
rmse_cdd = np.sqrt(mse_cdd)

mse_cdi = mean_squared_error(test_cdi, forecast_cdi_arma)
rmse_cdi = np.sqrt(mse_cdi)

# Même chose, on calcule les erreurs MSE et RMSE pour les secteurs 1, 2 et 3
mse_sec1 = mean_squared_error(test_sec1, forecast_sec1_arma)
rmse_sec1 = np.sqrt(mse_sec1)

mse_sec2 = mean_squared_error(test_sec2, forecast_sec2_arma)
rmse_sec2 = np.sqrt(mse_sec2)

mse_sec3 = mean_squared_error(test_sec3, forecast_sec3_arma)
rmse_sec3 = np.sqrt(mse_sec3)

# On crée un dico avec les erreurs dedans
error_ARAM = {
    'MSE': [mse_cdd, mse_cdi, mse_sec1, mse_sec2, mse_sec3],
    'RMSE': [rmse_cdd, rmse_cdi, rmse_sec1, rmse_sec2, rmse_sec3]
}

# On crée un dataframe des erreurs 
error_ARAM = pd.DataFrame(error_ARAM, index=['CDD', 'CDI', 'Selected Secteur 1', 'Selected Secteur 2', 'Selected Secteur 3'])

# Affichage du dataframe
error_ARAM
```

## Comparaison des erreurs entre le premier modèle et le modèle ARIMA

- Pour l'erreur MSE, on remarque que le modèle ARIMA a des erreurs moins grandes que le premier modèle pour la courbe du CDI, ce qui signifie que le poids des pires erreurs est moindre que dans le modèle précédent (4.43E+10 pour ARIMA contre 7.40E+10 pour le précédent modèle)
En revanche, pour les contrats CDI, c'est le modèle précédent qui reste plus performant que le modèle ARIMA, avec une MSE de 2.1E+10 contre presque 5E+10 pour le modèle ARIMA. La différence reste donc assez importante.
Le modèle ARIMA est plus performant pour prédire la tendance des CDD que des CDI !
De même pour la prédiction effectuée sur les trois secteurs, la seule prédiction où la MSE est moins importante dans le modèle ARIMA concerne celle effectuée sur le secteur 3.

Globalement, lorsque le modèle ARIMA effectue des erreurs, les écarts sont plus importants que pour le modèle précédent.

- Pour l'erreur RMSE, on remarque que, si effectivement le modèle commet plus d'erreurs que le modèle précédent pour la prédiction des CDI, et moins pour la prédiction des CDD, il est surprenant de constater que pour la prédiction sur les trois secteurs, les choses s'inversent : pour les secteurs 1 et 2, la RMSE est moindre pour ARIMA, et pour le secteur 3, ARIMA enregistre une plus grande RMSE.

Cela signifie que ARIMA commet plus d'erreurs de prédictions que le modèle précédent pour les CDI et le secteur 3, mais moins pour les autres éléments (la RMSE accorde un poids égal à toutes les erreurs, donc une RMSE plus élevée pour un modèle indique simplement une "quantité" d'erreurs plus importante).

Globalement, le modèle ARIMA commet moins d'erreurs.

Ainsi, un modèle n'est pas significativement meilleur que l'autre et le choix du modèle va fortement dépendre de la prédiction que l'on souhaite effectuer.
